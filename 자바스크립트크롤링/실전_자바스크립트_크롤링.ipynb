### 패킷 스니핑을 이용해서 20개 매물의 매물명, 매물가격, 크기, 층수, 방향 추출하기

```python
import requests
import re, json
from bs4 import BeautifulSoup
```

```python
url = "https://begincoding.pythonanywhere.com/api/articles?page=1&pageSize=20"
```

```python
res = requests.get(url)
```

```python
datas = json.loads(res.text)
datas
```

```python
for data in datas :
    name = data.get("article_name",'')
    price = data.get("article_deal_or_warrant_price",'')
    size = data.get("article_area_size",'')
    floor = data.get("article_floor",'')
    direction = data.get("article_direction",'')
    print(name, '\n', price, '\n', size, '\n', floor, '\n', direction)
    print('=' * 20)
```

### 20개 매물의 매물명, 매물가격, 크기, 층수, 방향, 상세내용, 부동산명, 부동산주소 추출

```python
import requests
import re, json
from bs4 import BeautifulSoup
```

```python
url = "https://begincoding.pythonanywhere.com/api/articles?page=1&pageSize=20"
```

```python
res = requests.get(url)
```

```python
datas = json.loads(res.text)
datas
```

```python
for data in datas :
    
    name = data.get("article_name",'')
    price = data.get("article_deal_or_warrant_price",'')
    size = data.get("article_area_size",'')
    floor = data.get("article_floor",'')
    direction = data.get("article_direction",'')
    
    num = data.get("article_no",'')
    sub_url = 'https://begincoding.pythonanywhere.com/level1/{}'.format(num)
    sub_res = requests.get(sub_url, 'html.parser')
    soup = BeautifulSoup(sub_res.text)
    text = soup.select_one('script').text
    pattern = r'window\.__PRELOADED_STATE__ = (\{.*\});'
    match = re.search(pattern, text, re.DOTALL)
    
    if match:
        json_text = match.group(1)
        sub_data = json.loads(json_text)
        content = sub_data.get('article_detail_desc')
        estate_name = sub_data.get('article_realtor_name')
        estate_address = sub_data.get('article_realtor_address')

    
    print(name, '\n', price, '\n', size, '\n', floor, '\n', direction, '\n')
    print('상세내용 = ',content, '\n\n','부동산 이름 = ' ,estate_name, '\n\n', '부동산 주소 = ',estate_address, '\n')
    print('=' * 20)
```
