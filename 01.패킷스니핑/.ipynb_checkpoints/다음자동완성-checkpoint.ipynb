```python
def parse_parameters(data):
    result_dict = {}
    lines = data.strip().split('\n')

    for line in lines:
        # ':'를 기준으로 파라미터 이름과 값을 분리
        if ':' in line:
            key, value = line.split(':', 1)
            key = key.strip()
            value = value.strip()
            # 값이 있는 경우만 딕셔너리에 추가
            if value:
                result_dict[key] = value

    return result_dict

# 예시 데이터
data_string = """
accept:
*/*
accept-encoding:
gzip, deflate, br, zstd
accept-language:
ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7
connection:
keep-alive
cookie:
webid=d675680d9a544559a15b205d88aa9f82; webid_ts=1739152109883; webid_sync=1742124418818; __gads=ID=c023626966c60766:T=1742123709:RT=1742124931:S=ALNI_MbnLwdGdS01QTNZ2O7ciPNdOmVjQQ; __gpi=UID=000010643f2511b7:T=1742123709:RT=1742124931:S=ALNI_MZqkoxcVK7IhyJa-pZhvpK-C2Xc5A; __eoi=ID=2468cdf21300a6cf:T=1742123709:RT=1742124931:S=AA-Afja6rYpt3yWekxRshlj7dsSV; _T_ANO=dlLbL7igMGqjQ7ErDxRkKr9ZDfEN1nlLBQiUMHQxzPUhswUK/VoQ1/91RuBqCmoPraCHlDcZq+3elJ0SLs3BZDXbwfM4Sx24AvgoKxLLnz0D/LtCDgu91RpTGSp2nZlqhCVPLOiBUr/CGsfXQJEKDMlFetXPbmozgvz9CROwqxWReCGJDJ/te2wx9MjuLlZqwsAM/NxgJJjm/j3/RYPKTrLLf7xgNF7yhrXS7GbEwVXbLJj8xlbLL0P0KWM50iLsT6B2goZHwoh+KLVRHFkkUE+I8sjdKjs+tA4cB05MpvsGoy4sb9AEpGbME0n1tJjqTBNsXsUve8prp0DyJ7t1MA==
host:
vmsuggest.search.daum.net
referer:
https://www.daum.net/
sec-ch-ua:
"Chromium";v="134", "Not:A-Brand";v="24", "Google Chrome";v="134"
sec-ch-ua-mobile:
?0
sec-ch-ua-platform:
"Windows"
sec-fetch-dest:
script
sec-fetch-mode:
no-cors
sec-fetch-site:
same-site
user-agent:
Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36
"""

# 함수 호출
result = parse_parameters(data_string)

# 결과 출력
for key, value in result.items():
    print(f"'{key}': '{value}',")
```

```python
import pyautogui
import time
import requests
import pandas as pd
import bs4
import re
from selenium.webdriver.support.select import Select
import json

sub_list = ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅅ', 'ㅇ', 'ㅈ', 'ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']
main_keyword = pyautogui.prompt('키워드를 입력해주세요')

result = []

for sub in sub_list :
    keyword = main_keyword + ' ' + sub
    url = "https://vmsuggest.search.daum.net/v2/sushi/pc/get?q={}".format(keyword)
    res = requests.get(url, headers = data_string)
    data_dict = res.json()
    data_dict['subkeys']
    for item in data_dict['subkeys'] :
        print(item['keyword'])
        result.append(item['keyword'])


df = pd.DataFrame(result, columns=['키워드'])
df.to_excel(f'다음자동완성_{main_keyword}.xlsx', index=False)
```
