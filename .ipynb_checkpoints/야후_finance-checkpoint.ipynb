### 국채수익률 데이터

```python
url = "https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value=2020"
```

```python
pd.read_html(requests.get(url).text)[0]
```

```python
cols_to_drop = [
    '20 YR', '30 YR', 'Extrapolation Factor',
    '6 WEEKS BANK DISCOUNT', 'COUPON EQUIVALENT', '8 WEEKS BANK DISCOUNT',
    'COUPON EQUIVALENT.1', '17 WEEKS BANK DISCOUNT', 'COUPON EQUIVALENT.2',
    '52 WEEKS BANK DISCOUNT', 'COUPON EQUIVALENT.3'
]
```

```python
df = pd.read_html(requests.get(url).text)[0].drop(columns=cols_to_drop, errors='ignore')
```

```python
df
```

```python
import time
df_list = []
for year in range(2010, 2021) : 
    url = "https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value={}".format(year)
    df_list.append(
        pd.read_html(requests.get(url).text)[0].drop(columns=cols_to_drop, errors='ignore')
    )
    time.sleep(0.5)
```

```python
df = pd.concat(df_list)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = df.set_index("Date")
df.head()
```

```python
df.plot(figsize = (20,10))
```
