```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from collections import defaultdict

# 1. 바코드 데이터 불러오기
df = pd.read_excel("라일락책_20250314.xlsx")
df['바코드'] = df['바코드'].apply(lambda x: x.replace('\xa0', '') if isinstance(x, str) else x)

# 2. 결과 저장용 딕셔너리
result_dict = defaultdict(list)

# 3. 세션 사용으로 속도 향상
session = requests.Session()

# 4. 바코드 순회
for num, code in enumerate(df['바코드'], start=1):
    print(f"[{num}/{len(df)}] 바코드: {code}")
    
    try:
        # 검색 결과 페이지 요청
        search_url = f"https://www.aladin.co.kr/search/wsearchresult.aspx?SearchTarget=Used&SearchWord={code}"
        res = session.get(search_url)
        soup = BeautifulSoup(res.text, "html.parser")

        # 상품 URL 추출
        code_url_table = soup.select_one("#Search3_Result > div:nth-child(1) > table")
        a_tag = code_url_table.select_one("div.ss_book_list > ul > li > a")
        if a_tag is None:
            continue

        domain = "https://www.aladin.co.kr"
        full_url = domain + a_tag['href']
    except Exception as e:
        print(f"❌ 검색 결과 파싱 실패: {e}")
        continue

    # 페이지 순회
    page = 1
    while True:
        print(f"  📄 페이지 {page} 요청 중...")

        try:
            detail_url = (
                f"{full_url}&TabType=0&QualityType=0&Fix=1&ShopCode=0"
                f"&HalfDelivery=0&SortOrder=9&page={page}&PublishDay=84"
                f"&PriceFilterMin=0&PriceFilterMax=0"
            )
            detail_res = session.get(detail_url)
            soup = BeautifulSoup(detail_res.text, "html.parser")

            # 중고 판매 리스트 테이블 추출
            detail_td = soup.select_one(
                "#Ere_prod_allwrap_box > div.Ere_prod_middlewrap > div.Ere_usedsell_table table > tbody"
            )
            rows = detail_td.select("tr")
            if len(rows) <= 1:
                break  # 데이터가 없으면 종료
        except Exception as e:
            print(f"  ❌ 페이지 파싱 실패: {e}")
            break

        # 각 중고 상품 정보 추출
        for row in rows[1:]:  # 첫 tr은 헤더이므로 제외
            try:
                book_name = row.select_one(".sell_tableCF2 a").text.strip()
                book_grade = row.select_one("span.Ere_sub_top").text.strip()
                price = row.select_one(".Ere_sub_pink").text.strip()
                delivery = [li.text.strip() for li in row.select(".price li")[2:]]
                
                result_dict[book_name].append({
                    'book_name': book_name,
                    'book_grade': book_grade,
                    'price': price,
                    'delivery': ', '.join(delivery)
                })

                print(f"    📚 {book_name} | {book_grade} | {price} | {delivery}")

            except Exception as e:
                print(f"    ⚠️ 행 파싱 오류: {e}")
                continue

        page += 1

# 5. 결과 정리 및 저장
all_rows = []
for book_group in result_dict.values():
    all_rows.extend(book_group)

df_result = pd.DataFrame(all_rows)
df_result.to_csv("알라딘_도서_중고정보.csv", index=False, encoding="utf-8-sig")
print("✅ CSV 저장 완료: 알라딘_도서_중고정보.csv")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from collections import defaultdict

# 1. 데이터 불러오기
df = pd.read_excel("라일락책_20250314.xlsx")
df['바코드'] = df['바코드'].apply(lambda x: x.replace('\xa0', '') if isinstance(x, str) else x)

# 2. 등급 매핑
grade_map = {'최상': 1, '상': 2, '중': 3}

# 3. 결과 저장용 딕셔너리
result_dict = defaultdict(list)

# 4. requests 세션 사용
session = requests.Session()

# 5. 바코드 + 품질등급 순회
for num, (code, want_grade_str) in enumerate(zip(df['바코드'], df['품질등급']), start=1):
    print(f"\n[{num}/{len(df)}] 바코드: {code} | 등급: {want_grade_str}")

    want_grade = grade_map.get(str(want_grade_str).strip(), 0)

    # 검색 페이지
    try:
        search_url = f"https://www.aladin.co.kr/search/wsearchresult.aspx?SearchTarget=Used&SearchWord={code}"
        res = session.get(search_url)
        soup = BeautifulSoup(res.text, "html.parser")
        code_url_table = soup.select_one("#Search3_Result > div:nth-child(1) > table")
        a_tag = code_url_table.select_one("div.ss_book_list > ul > li > a")
        if a_tag is None:
            continue
        full_url = "https://www.aladin.co.kr" + a_tag['href']
    except Exception as e:
        print(f"❌ 검색 오류: {e}")
        continue

    # 6. 두 번 크롤링: 전체등급 / 등급필터
    for mode, quality_type in [('전체', 0), ('등급필터', want_grade)]:
        print(f"  🔍 모드: {mode} | QualityType={quality_type}")
        page = 1
        while True:
            try:
                detail_url = (
                    f"{full_url}&TabType=0&QualityType={quality_type}&Fix=1&ShopCode=0"
                    f"&HalfDelivery=0&SortOrder=9&page={page}&PublishDay=84"
                    f"&PriceFilterMin=0&PriceFilterMax=0"
                )
                res = session.get(detail_url)
                soup = BeautifulSoup(res.text, "html.parser")
                detail_td = soup.select_one(
                    "#Ere_prod_allwrap_box > div.Ere_prod_middlewrap > div.Ere_usedsell_table table > tbody"
                )
                rows = detail_td.select("tr")
                if len(rows) <= 1:
                    break
            except Exception as e:
                print(f"  ❌ 페이지 파싱 오류: {e}")
                break

            for row in rows[1:]:
                try:
                    book_name = row.select_one(".sell_tableCF2 a").text.strip()
                    book_grade = row.select_one("span.Ere_sub_top").text.strip()
                    price = row.select_one(".Ere_sub_pink").text.strip()
                    delivery = [li.text.strip() for li in row.select(".price li")[2:]]

                    result_dict[book_name].append({
                        'book_name': book_name,
                        'book_grade': book_grade,
                        'price': price,
                        'delivery': ', '.join(delivery),
                        'mode': mode
                    })

                    print(f"    📚 {book_name} | {book_grade} | {price} | {delivery}")
                    break  # 최저가만 추출

                except Exception as e:
                    print(f"    ⚠️ 행 파싱 오류: {e}")
                    continue

            break  # 페이지 1페이지만 확인 (최저가용)

# 7. 결과 정리 및 저장
all_rows = []
for book_group in result_dict.values():
    all_rows.extend(book_group)

df_result = pd.DataFrame(all_rows)
df_result.to_csv("알라딘_도서_중고정보.csv", index=False, encoding="utf-8-sig")
print("\n✅ CSV 저장 완료: 알라딘_도서_중고정보.csv")
```
