# Session 방식

```python
import requests

# 세션 객체 생성
session = requests.Session()

login_data = {
    'username': 'test4',
    'password': '4444',
    'submit': '로그인',
}

# Content-Type: application/x-www-form-urlencoded면 → data=... 사용해야 합니다.

# Content-Type: application/json이면 → json=... 사용합니다.


res = session.post("https://begincoding.pythonanywhere.com/login", data = login_data)
print(res.text)
```

```python
comment_data = {
    'content': '긍정적인 마음을 가질 수 있음에 감사합니다',
    'submit': '등록하기',
}



res = session.post("https://begincoding.pythonanywhere.com/level2/comment/new", data = comment_data)
print(res.text)
```

# 로그인이 어려울때 (고급 팁)

### 1. 셀레니움으로 로그인을 먼저 진행

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("http://begincoding.pythonanywhere.com/login")
time.sleep(1)

username = driver.find_element(By.CSS_SELECTOR, "#username")
username.send_keys('test4')
time.sleep(1)

password = driver.find_element(By.CSS_SELECTOR, "#password")
password.send_keys('4444')
time.sleep(1)

login_btn = driver.find_element(By.CSS_SELECTOR, "#submit")
login_btn.click()
```

### 2. 브라우저의 쿠키값을 그대로 가져옴

```python
cookies = driver.get_cookies()
print(cookies)
```

### 3. 세션 객체를 만들고 가져온 쿠키를 추가

```python
session = requests.Session()

for cookie in cookies:
    session.cookies.set(cookie['name'], cookie['value'])
```

### 4. 로그인이 필요한 페이지를 요청

```python
comment_data = {
    'content': '오늘도 지식이 늘어남에 감사합니다',
    'submit': '등록하기',
}



res = session.post("https://begincoding.pythonanywhere.com/level2/comment/new", data = comment_data)
print(res.text)
```

```python
from bs4 import BeautifulSoup

# 세션 객체 정의
session = requests.Session()

login_data = {
    'username' : 'test4',
    'password' : '4444',
    'submit' : '로그인',
}

# 로그인을 하고 동시에 res에는 세션이 존재하는 페이지 반환
res = session.post('http://begincoding.pythonanywhere.com/login', data = login_data)

# 로그인이 유지 안되어있으면 로그아웃 이라는 텍스트가 없으니까 로그인이 되어있을때만 아래 실행
if "로그아웃" in res.text :

    # 세션을 가지고 댓글 수정할 페이지로 진입
    res = session.get('https://begincoding.pythonanywhere.com/level2?page=2')
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')

    # 해당 세션과 쌍을 맺는 csrf토큰을 가져옴
    # attrs은 해당 태그에서 속성값을 가져오는데 이 코드에서는 value라는 속성의 값을 가져옴
    csrf_token = soup.select_one("#csrf_token").attrs['value']

    # post로 넘겨줄 데이터를 정의
    comment_data = {
        'csrf_token' : csrf_token, # csrf토큰 
        'content' : '모레도 지식이 늘어남에 감사합니다', # 수정할 내용
        'submit' : '수정' # 수정 버튼
    }

    # 2페이지에 있는 22번 댓글에 comment_data로 넘겨줌
    res = session.post('https://begincoding.pythonanywhere.com/level2/comment/22/edit?page=2', comment_data)
    print(res.text)
```
